```python
import random

chains = [1] * 10000
for _ in range(0, 4000):
    i1, i2 = random.choices(range(len(chains)), weights=chains, k=2)
    if i1 == i2: continue
    chains[i1] += chains[i2]
    del chains[i2]
```

```python
%matplotlib inline
from plot import plot

xs = range(0, len(chains))
ys = sorted(chains, reverse=True)
plot('Paperclips', xs, ys, xlabel='chains', ylabel='chain length')
```
