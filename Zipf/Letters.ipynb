```python
ranks = {chr(0x61 + i): 0 for i in range(0, 26)}
with open('big.txt') as file:
    for char in file.read():
        if char in ranks:
            ranks[char] = ranks[char] + 1
```

```python
%matplotlib inline
from plot import plot

xs = range(0, len(ranks))
ys = sorted(ranks.values(), reverse=True)
plot('Letters', xs, ys, xticks=xs, xticklabels=ranks.keys(), ylabel='# of uses')
```
