```python
import random
letters = "abc "
random_text_len = 1024 * 1024
random_text = "".join([random.choice(letters) for _ in range(1, random_text_len)])
```

```python
ranks = {}
for word in random_text.split():
    ranks[word] = ranks.get(word, 0) + 1
```

```python
%matplotlib inline
from plot import plot

xs = [rank for word, rank in ranks.items()]
ys = [len(word) / random_text_len for word, rank in ranks.items()]

plot('Random Text', xs, ys, xlabel='word rank', ylabel='% of text')
```
