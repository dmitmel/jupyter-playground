```python
from graphviz import Digraph
import matplotlib.pyplot as plt
```

```python
HEIGHT = 10

rows = [[0] * (y+1) for y in range(0, HEIGHT)]
rows[0][0] = 1

for y in range(1, HEIGHT):
    for x in range(0, y + 1):
        left = rows[y - 1][x - 1] if x > 0 else 0
        right = rows[y - 1][x] if x < y else 0
        rows[y][x] = left + right
```

```python
dot = Digraph(comment="Pascal Triangle")

def node_name(x, y): return "{} {}".format(x, y)

for y, row in enumerate(rows):
    for x, node in enumerate(row):
        dot.node(node_name(x, y), str(node))

for y, row in enumerate(rows[:-1]):
    for x, node in enumerate(row):
        dot.edge(node_name(x, y), node_name(x, y + 1))
        dot.edge(node_name(x, y), node_name(x + 1, y + 1))

dot
```
